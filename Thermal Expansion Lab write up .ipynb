### Introduction
The coefficient of linear expansion, $\alpha$, for the metals copper and steel was determined experimentally by heating a metal pipe, of copper and steel. The variables that were acquired by during the experiment were, recording room temperature, $T_i$. Then, heating up a metal pipe with steam from boiling water, measuring the amount the pipe grew, $\Delta L$, via a dial. Then, recording the final temperature of the pipe, $T_f$, via a resister and utilizing a computer program to convert kiloohms to degrees Celsius.
An objects expansion due to heat is given by, $\alpha = \frac{\Delta L}{L\Delta T}$

Where $\Delta L$ is the amount the metal tube expanded during the experiment, via the dial, measured in meters. $L$ is the initial length of the tube, measured in meters via a meter stick. $\Delta T$ is the temperature difference $Temperature_i-Temperature_f$ measured by the resister and then converted to degrees Celsius via a computer program.
